```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('D:/Machine Learning/Feature Scaling/Standardization/tips.csv')
```

```python
df.head()
```

```python
import numpy as np
mean=np.mean(df['total_bill'])
std=np.std(df['total_bill'])
print(mean,std)
```

```python
standardized_data=[]
for i in list(df['total_bill']):
    z_score=(i-mean)/std
    standardized_data.append(z_score)
```

```python
standardized_data
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(standardized_data)
```

### Standardization using StandardScalar

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill']]),columns=['total_bill'])
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]),columns=['total_bill','tip'])
```
